```python
#uploaded = files.upload()
#df = pd.read_json(io.StringIO(uploaded.get('who_nmr2.json').decode('utf-8')))
df = pd.read_json('who_nmr2.json')
df_2021 = df[df['year'] == 2021]
df_2021
```

```python
import altair as alt
import pandas as pd
from vega_datasets import data

source = alt.topo_feature(data.world_110m.url, "countries")

background = alt.Chart(source).mark_geoshape(fill="white")

foreground = (
    alt.Chart(source).mark_geoshape(
        stroke="black", strokeWidth=0.2
    ).encode(
        color=alt.Color(
            "NMR:N", scale=alt.Scale(scheme="lightgreyred"), legend=None,
        ),
        tooltip=[
            alt.Tooltip("Country:N", title="Country"),
            alt.Tooltip("NMR:Q", title="Neonatal deaths per 1k live births"),
        ],
    ).transform_lookup(
        lookup="id",
        from_=alt.LookupData(df_2021, "NumericCode", ["NMR", "Country"]),
    )
)

chart = (
    (background + foreground)
    .configure_view(strokeWidth=0)
    .properties(width=700, height=400)
    .project("naturalEarth1")
)

chart
```
