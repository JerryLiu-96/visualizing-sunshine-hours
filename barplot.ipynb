```python
from matplotlib.colors import LinearSegmentedColormap
# Define a gradient color map
cmap = LinearSegmentedColormap.from_list('mycmap', [(0, '#feedde'), (0.25, '#fdbe85'), (0.5,'#fd8d3c'),(0.75,'#e6550d'),(1, '#a63603')])
```

```python
330*0.75
```

```python
# create a dataset
height = np.array(df['sunshine'][0:12])
month = np.array(df['month'][0:12])
x_pos = np.arange(len(month))

# Create the bar plot with gradient colors
fig, ax = plt.subplots()
bars = ax.bar(month, height, color=cmap(height / max_sunshine))

# Create names on the x-axis
plt.xticks(x_pos, month)
plt.title('Average Hours of Sunshine in Chicago')
plt.xlabel('Month')
plt.ylabel('Hours of Sunshine')
plt.ylim(0, 330)

# Add a color bar to show the mapping of values to colors
cbar=fig.colorbar(plt.cm.ScalarMappable(norm=None, cmap=cmap), ax=ax)
cbar.set_ticks([0, 0.25,0.5,0.75, 1.0])
cbar.set_ticklabels(['0', '82.5','155','247.5', '330'])

# Show graph

fig.set_dpi(300)

plt.savefig("img/output_chi.png",format='png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap
# Define a gradient color map
cmap = LinearSegmentedColormap.from_list('mycmap', [(0, '#deebf7'), (0.5, '#9ecae1'), (1, '#3182bd')])

# Generate some random data for the bar plot
x = np.arange(10)
y = np.random.rand(10) * 10

# Create the bar plot with gradient colors
fig, ax = plt.subplots()
bars = ax.bar(x, y, color=cmap(y / max(y)))

# Add a color bar to show the mapping of values to colors
cbar=fig.colorbar(plt.cm.ScalarMappable(norm=None, cmap=cmap), ax=ax)
cbar.set_ticks([0, 0.5, 1.0])
cbar.set_ticklabels(['Low', 'Medium', 'High'])

plt.show()
```
